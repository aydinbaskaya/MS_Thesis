### Power Equations for Wind Energy

The available shaft power(output power) $P$ from a wind turbine can be expressed as a function of the wind speed as follows :

## $$P = \frac{1}{2}\rho_{air}C_{p}(\lambda,\beta)\pi r^2v^3$$
where,

$\rho_{air}$ is the mass density of air

$C_{p}(\lambda,\beta)$ is the power coefficient, a function of the tip speed ratio $\lambda$ and the pitch angle $\beta$


$r$ is the blade length

$v$ is the wind velocity




#### Power Coefficient

Power coefficient(sometimes called performance coefficient) is basically can be defined as the ratio between captured wind power by the wind turbine and the available input power of the wind.Therefore it tells us **how efficiently** we utilize the wind turbine.Its value is sometimes taken from look up tables or can be assumed by nonlinear computations.[25] 

In [19](http://ieeexplore.ieee.org/stamp/stamp.jsp?tp=&arnumber=6226866&isnumber=6331663) Power coefficient $C_{p}$ is defined  as a nonlinear function of TSR($\lambda$) and pitch angle $\beta$ as follows,

### $$C_{p}=c_{1}(\frac{c_{2}}{\lambda_{i}}-c_{3}\beta-c_{4})e^{\frac{-c_{5}}{\lambda_{i}}}+c_{6}\lambda$$

where

### $$\frac{1}{\lambda_{i}}=\frac{1}{\lambda+0.08\beta}-\frac{0.035}{\beta^3+1}$$

The coefficients(depends on the turbine characteristics):

$c_{1}$=0.5176

$c_{2}$=116

$c_{3}$=0,4

$c_{4}$=5

$c_{5}$=21

$c_{6}$=0,0068

#### Tip Speed Ratio

* Tip speed ratio $\lambda$ (TSR) is defined as a ratio of tip speed of turbine blade to speed of the wind. This ratio is very useful when designing a wind turbine.

$$\lambda = \frac{R\omega_{M}}{V} $$

where,
$V$ is the wind speed

$\omega_{M}$ is the rotational rotor speed

$R$ is the rotor radius



* TSR is a kind of measurement of how speedy turbine blades turn. This topic is very important because of the efficiency issues. If turbine blades turns too slow then incoming air to the turbine is not used efficiently (Remember Betz limit!). If turbine blades turns too fast, blades act like solid wall to the turbine and then efficiency decreases again.To avoid bad consequences(low efficiency, physical breakdown) of turbulence issue,choosing optimal TSR is really important.

#### Betz Limit
* Theoretically maximum **59%** (theoretical maximal of $C_{p}$ ) of energy carried by the wind can be extracted by a wind turbine. This result is concluded by German physicist **Albert Betz**
* Generally imperfections and in blade manufacture reduces the actual energy yield of the turbine less than the useable energy.

#### Weibull Distribution and other issues
* Time dependent nature of the wind determines the production scheme of the WECs(Wind Energy Conversion Systems). There are some approaches for estimate the wind profile at given place. Wind profile at any given area is generally measured 10 meters above the ground level and estimated by **1/7 Power Profile Law** as given follows[25] : 

### $$\frac{v}{v_{0}}=(\frac{h}{h_{0}})^{\alpha(v_{0},h_{0})}$$

where, 

$v$ is desired wind speed,

$v_{0}$ is reference wind speed, 

$h$ is the desired height,

$h_{0}$ is the reference height 


$\alpha$ value in the equation above can be calculated as follows[25],

### $$\alpha(v_{0},h_{0})=\frac{0.37-0.088\ln(v_{0})}{1-0.088\ln(\frac{h}{h_{0}})}$$

this value is approximated as **"1/7"** in calculations.









* One remote sensing tecnique of the wind profile is **LIDAR(Light Detection and Ranging)** which is based on Doppler Effect to determine wind characteristics. In another method, **ANFIS(Adaptive Neuro-fuzzy interference system)** based clustering algorithm is used to predict wind speed. 


* **Weibull distribution** is used to determine the wind distribution and gives an indication of what percentage of time a certain wind speed occurs in a given site. Weibull distribution is given as follows:

## $$f(v)=\frac{k}{a} (\frac{v}{a})^{(k-1)}e^{-(\frac{v}{a})^k}$$
where,

$v$ is the wind speed

$k$ is the shape parameter

$a$ is the scale parameter

* Therefore energy yield over a period of time $T$ can be defined as follows,

$$T\int P(v)f(v)dv$$

where,
 P is the generated power 
 
 f(v) is the Weibull function
 
 T=8760 hours (1 year)



* Minimum wind speed that is needed to start to rotate the turbines' blade is **cut-in wind speed**

**Note:** mass with the *m* indice denotes turbine moment of inertia and other mass with "g" indice denotes generator moment of inertia.

* Equations for angular speeds at the both end of the two-mass model are as follows[26],

$$\frac{d\omega_{m}}{dt}=\frac{1}{J_{m}}(T_{m}-T_{dt}-T_{at}-T_{ts})$$ <br><br>
$$\frac{d\omega_{g}}{dt}=\frac{1}{J_{g}}(T_{ts}-T_{dg}-T_{ag}-T_{g})$$<br>
where,<br><br>
$J_{m}$ is moment of inertia of blades and hub<br><br>
$T_{m}$ is mechanical torque,<br><br>
$T_{dt}$ is resistant torque in the wind turbine bearing<br><br>
$T_{at}$ is resistant torque in the hub and blades <br><br>
$J_{g}$ is moment of inertia of rotor of the generator<br><br>
$T_{g}$ is electrical torque,<br><br>
$T_{dg}$ is resistant torque in generator bearing<br><br>
$T_{ag}$ is resistant torque for airflow in generator <br><br>
$T_{ts}$ is torsional stifness torque<br><br>






* In some European countries, there is limited space for installing wind turbines onshore, so there is a increasing trend for install them offshore. However, installing these wind turbines offshore requires more expensive and frequent maintenance than that of onshore. Therefore direct-drive technology gains importance by its robust and simple drive-train and easy-to-install structure in practice.


* Direct-drive(DD) wind turbine concept eliminates multi-stage gearbox which means extra cost of maintenance and extra mechanical cost in variable speed constant frequency(VSCF) type wind turbines


* According to [Laithwaite](https://en.wikipedia.org/wiki/Eric_Laithwaite)'s and [Freris](http://www.lboro.ac.uk/departments/eese/staff/leon-freris/)'s work, power output of any rotating electrical machine can be generally desribed by the formula given below: (which is rough formula than given previous one) 

$$P=KD^2Ln$$

where,

D is rotor diameter

L is length(axial)

n is rotational speed

K is constant

* Note that output power is related to airgap diameter squared,therefore DD generators have larger diameters. 

* Since DD's have no gear-box, n is very small. Therefore,it's inevitable to increase diameter or axial length.Usually increasing diameter is more preferable since it affects the output power by square of resulting value.




#### Flux orientations

One of the important options in design of the machine is magnetic flux path directions. Current technology in this topic can be classied as three main subgroups as follows:

* Radial flux (path of the magnetic flux is perpendicular to the direction of the rotor shaft)

* Axial flux (path of the magnetic flux is perpendicular to the radial direction of the rotor shaft)

* Transverse flux (path of the magnetic flux is perpendicular to the direction of the rotor rotation)

#### Grid connection and power electronic issues

* Since C-GEN concept generators has no excitation windings(instead there are permanent magnets), output voltage is dependent on machine speed. Therefore, proper power electronics control circuits are needed to make efficient grid connection. 

* WECs based on power electronic converters(PEC) utilizes two different converter circuits and their control blocks. Their names are as follows

    $a)$ RSC(rotor side converter) ~generally used in DFIG
    
    $b)$ GSC(grid side converter)
    

* There are three main schemes to be used in these circuit topologies:

a) IGBT Supply side inverter

* In this topology, generator output AC voltage is rectified with six diode bridge(uncontrolled) rectifier. Rectified uncontrolled voltage directly connected to DC-link capacitor.

* Then this rectified voltages is converted again to AC voltage via inverter.

* This topology has lower cost than other options since there is no controlled rectifier unit.

b) Intermediate DC/DC Converter 

* Biggest advantage of this topology is its cost-benefit especially when working with multiple turbine systems.

* Instead of using uncontrolled and bulk dc-link capacitor, a DC chopper is used after rectifier unit.

* By this chopper controlled inverter has more smooth and flexible control on grid connection. Also dc-link capacitors are smaller in size.

* Maximum power point tracker algorithm can be used to drive chopper control(duty cycle) according to variable output power of the generator.
