The available shaft power(output power) $P$ from a wind turbine can be expressed as a function of the wind speed as follows :

##$$P = \frac{1}{2}\rho_{air}C_{p}(\lambda,\beta)\pi r^2v^3$$
where,

$\rho_{air}$ is the mass density of air

$C_{p}(\lambda,\beta)$ is the power coefficient, a function of the tip speed ratio $\lambda$ and the pitch angle $\beta$





####Tip Speed Ratio

* Tip speed ratio $\lambda$ (TSR) is defined as a ratio of tip speed of turbine blade to speed of the wind. This ratio is very useful when designing a wind turbine.

####$$\lambda = \frac{R\omega_{M}}{V} $$

where,
$V$ is the wind speed

$\omega_{M}$ is the rotational rotor speed

$R$ is the rotor radius



* TSR is a kind of measurement of how speedy turbine blades turn. This topic is very important because of the efficiency issues. If turbine blades turns too slow then incoming air to the turbine is not used efficiently (Remember Betz limit!). If turbine blades turns too fast, blades act like solid wall to the turbine and then efficiency decreases again.To avoid bad consequences(low efficiency, physical breakdown) of turbulence issue,choosing optimal TSR is really important.

####Power Coefficient

Power coefficient(sometimes called performance coefficient) is basically can be defined as the ratio between captured wind power by the wind turbine and the available input power of the wind.Therefore it tells us how efficiently we utilize the wind turbine.

####Betz Limit
* Theoretically maximum **59%** of energy carried by the wind can be extracted by a wind turbine. This result is concluded by German physicist **Albert Betz**
* Generally imperfections and in blade manufacture reduces the actual energy yield of the turbine less than the useable energy.

####Weibull Distribution and other issues
* **Weibull distribution** is used to determine the wind distribution and gives an indication of what percentage of time a certain wind speed occurs in a given site. Weibull distribution is given as follows:

##$$f(v)=\frac{k}{a} (\frac{v}{a})^{(k-1)}e^{-(\frac{v}{a})^k}$$
where,

$v$ is the wind speed

$k$ is the shape parameter

$a$ is the scale parameter

* Therefore energy yield over a period of time $T$ can be defined as follows,

$$T\int P(v)f(v)dv$$

where,
 P is the generated power 
 
 f(v) is the Weibull function
 
 T=8760 hours (1 year)



Minimum wind speed that is needed to start to rotate the turbines' blade is **cut-in wind speed**


* In some European countries, there is limited space for installing wind turbines onshore, so there is a increasing trend for install them offshore. However, installing these wind turbines offshore requires more expensive and frequent maintenance than that of onshore. Therefore direct-drive technology gains importance by its robust and simple drive-train and easy-to-install structure in practice.


* Direct-drive(DD) wind turbine concept eliminates multi-stage gearbox which means extra cost of maintenance and extra mechanical cost in variable speed constant frequency(VSCF) type wind turbines


* According to [Laithwaite](https://en.wikipedia.org/wiki/Eric_Laithwaite)'s and [Freris](http://www.lboro.ac.uk/departments/eese/staff/leon-freris/)'s work, power output of any rotating electrical machine can be generally desribed by the formula given below: (which is rough formula than given previous one) 

$$P=KD^2Ln$$

where,

D is rotor diameter

L is length(axial)

n is rotational speed

K is constant

* Since DD's have no gear-box, n is very small. Therefore,it's inevitable to increase diameter or axial length.Usually increasing diameter is more preferable since it affects the output power by square of resulting value.




#### Flux orientations

One of the important options in design of the machine is magnetic flux path directions. Used technology in this topic can be classied as three main subgroups as follows:

* Radial flux (path of the magnetic flux is perpendicular to the direction of the rotor shaft)

* Axial flux (path of the magnetic flux is perpendicular to the radial direction of the rotor shaft)

* Transverse flux (path of the magnetic flux is perpendicular to the direction of the rotor rotation)
