##Direct-Drive Wind Turbines

* Main purposes of the Direct-drive(Ddrive) concept for wind turbines are;increasing energy yield,reducing gearbox failures and lower maintenence problems.

* Due to tip speed limitation they are operating at low speeds which is a important advantage. But according to the formula:

$$P=T.W_{m}$$

Torque must be increase inverse proportion to angular speed.

* To do that when scaling up the turbine sizes, amount of material is increased in order to maintain the air-gap against deflection forces between stator and rotor parts. This means DDrive machines are heavier than other types of machines. **Torque/volume** and **mass/torque** parameters are important during the design of the generator system. 
* More material also means extra cost which is big disadvantage for this type of generators.
* In scope of this notebook you can see different types of machine topologies. So there are some abbreviations given below:

*DFIG 3G* - Doubly-fed induction generator system with 3 stage gearbox

*PMSG DD* - Direct-drive permanent magnet synchronous generator

*PMSG 1G* - Permanent magnet synchronous generator with 1 stage gearbox

*PMSG DD* - Direct-drive permanent magnet synchronous generator

*EESG DD* - Direct-drive electrically excited synchronous generator

####C-GEN Concept

- C-GEN concept is a new generator topology developed by University of Edinburgh

- This technology consist of axial flux air-cored D-Drive PM generator 

- Main advantages of this concept among other technologies can be defined as follows:

    * There is no attraction force between stator and rotor due to air-cored structure
    * There is no cogging torque
    * Modular structure (identical machines can be stacked in order to increase power output)
    * These stacked generator sets can be grouped in parallel fashion, thus overall **reliability** increases.
    * Minumum structural mass due to reduced active material 
    * Generator is axial-flux type so axial length is minumum among other machine options
    

C-GEN concept was applied to different areas of renewable energy sources such as wind and marine energy. You can see the one of the C-GEN concept designed generator picture below

a)Electromagnetic design

We need some dimensions of the machine in order to calculate flux density in the air gap and flux linkage

```python
from IPython.display import Image
Image(filename='coilnmagnet.png')
```

We assume air gap flux density distribution is sinusoidal. Therefore,

$$B(\theta)= \hat{B}.\cos{(\frac{\pi \theta}{\theta_{p}})}  \ \  \ \ \ \     for \ R_{i} < r < R_{o}  \\\theta_{p}= \frac{2\pi}{N_{p}}$$

for the flux linkage,

$$\hat{\lambda}=\frac{4.\hat{B}.N_{t}.(R_{o}^{2}-R_{i}^{2})}{N_{p}^{2}(\theta_{o}-\theta_{i})}\Bigg(\cos(\frac{N_{p}\theta_{i}}{2})-\cos(\frac{N_{p}\theta_{o}}{2})\Bigg)$$

Assume sinusiodal flux linkage.Therefore,

$$\lambda= \hat{\lambda}.\sin(wt)$$


where $N_{t}$ is the number of turn of coil and $N_{p}$ is the number of poles

Now emf per coil can be computed by using [Faraday's Law](http://hyperphysics.phy-astr.gsu.edu/hbase/electric/farlaw.html) , namely,

$$emf= -N\frac{\Delta B.A}{\Delta t}$$

Voltage, current and efficiency values can be calculated according to these emf values. But remember that resistance values are changing during the various temperature conditions as follows,

$$R_{new}=R_{amb}(1+\alpha(T_{new}-T_{amb}))$$

where,

$R_{new}$ and $T_{new}$ are the resistance and temperature values of the current state, respectively

$R_{amb}$ and $T_{amb}$ are the resistance and temperature values of the previous state, respectively

$\alpha$ is the temperature coefficient


So we need a iterative optimization technique during the design process of these varying parameters. For these purposes, Genetic Algorithm Optimization(can be abbreviated as GAO) tecniques gives not most efficient or lightest but **optimum** solutions. This technique and its algorithm will be considered next days. 





To sum up, an electromagnetic design should cover following specifications:

* Number of phases
* Number of poles
* Number of coils
* Number of Turns
* Phase inductance and resistance
* Conductor area
* Air-gap flux density

c)Thermal Design 

Thermal performance of the machine effects and limits all electromagnetic designs. Therefore thermal cooling schemes are very important. 

* There are three main strategies about cooling issues of electrical machines;

1) Natural air/water cooling

2) Forced air/water cooling

3) Tidal current as cooling fluid(if possible)

####Grid connection and power electronic issues

* Since C-GEN concept generators has no excitation windings(instead there are permanent magnets), output voltage is dependent on machine speed. Therefore, proper power electronics control circuits are needed to make efficient grid connection. 

* There are three main schemes to be used in these circuit topologies:

a) IGBT Supply side inverter

* In this topology, generator output AC voltage is rectified with six diode bridge(uncontrolled) rectifier. Rectified uncontrolled voltage directly connected to DC-link capacitor.

* Then this rectified voltages is converted again to AC voltage via inverter.

* This topology has lower cost than other options since there is no controlled rectifier unit.

Block diagram of this topology is given below:

```python
from IPython.display import Image
Image(filename='igbtsupply.png')
```

*IGBT Supply side inverter option*

b) Intermediate DC/DC Converter 

* Biggest advantage of this topology is its cost-benefit especially when working with multiple turbine systems.

* Instead of using uncontrolled and bulk dc-link capacitor, a DC chopper is used after rectifier unit.

* By this chopper controlled inverter has more smooth and flexible control on grid connection. Also dc-link capacitors are smaller in size.

* Maximum power point tracker algorithm can be used to drive chopper control(duty cycle) according to variable output power of the generator.  

Block diagram of this topology is given below:

```python
from IPython.display import Image
Image(filename='intermediate.png')
```

*Intermediate dc/dc converter option*

c) Back-to-back PWM converter

* Intead of using separate chopper and rectifier units, only one AC/DC circuit block can be used.

* This option can require more complex control issues than other options.

Block diagram of this topology is given below:

```python
from IPython.display import Image
Image(filename='backtoback.png')
```

*Back-to-back converter option*

* When combining stacked generators into grid its important to choose number of generators in a group. Although number of a generator in a group is changed depend o the whole system, it's useful to group them in small number of parallel groups in order to maintain electricity generation during a fault condition.


* Remember connection type can effect the equipment and overall efficiency:

  - As number of parallel connected machine increases, dc-link current increases
