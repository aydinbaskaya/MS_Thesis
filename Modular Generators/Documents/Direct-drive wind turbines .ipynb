##Direct-Drive Wind Turbines

* Main purposes of the Direct-drive(Ddrive) concept for wind turbines are;increasing energy yield,reducing gearbox failures and lower maintenence problems.

* Due to tip speed limitation they are operating at low speeds which is a important advantage. But according to the formula:

$$P=T.W_{m}$$

Torque must be increase inverse proportion to angular speed.

* To do that when scaling up the turbine sizes, amount of material is increased in order to maintain the air-gap against deflection forces between stator and rotor parts. This means DDrive machines are heavier than other types of machines. **Torque/volume** and **mass/torque** parameters are important during the design of the generator system. 
* More material also means extra cost which is big disadvantage for this type of generators.
* In scope of this notebook you can see different types of machine topologies. So there are some abbreviations given below:

*DFIG 3G* - Doubly-fed induction generator system with 3 stage gearbox

*PMSG DD* - Direct-drive permanent magnet synchronous generator

*PMSG 1G* - Permanent magnet synchronous generator with 1 stage gearbox

*PMSG DD* - Direct-drive permanent magnet synchronous generator

*EESG DD* - Direct-drive electrically excited synchronous generator

* Ddrive concept thus is usually designed with a large diameter and small pole pitch to increase the efficiency, to reduce the active material(electromagnetic structure) and to keep the end winding losses small.

* Active material refers to electromagnetic structure while inactive part means mechanical structure

####PM Machine Structures for DDrive Concept

*RFPM*  -Radial Flux PM Machine

*AFPM**  -Axial Flux PM Machine

*TFPM*  -Transverse Flux PM Machine

*Among these topologies AFPM Machine can be studied for my thesis.

* DDrive PM generator system has the highest performance and reliability compared to both geared generator system and the direct-drive EESG according to my references.

####Improving performance

* Reduce both active and inactive material in order to get higher efficiencies
* **Modular construction** to get better assemble and disassemble(maintain and repair) conditions. This concept also allows each module can work individually thus better **fault-tolerance**
* PMSG DD has the highest energy yield
* Usually easy production and transportation is more preferable

Power output of this machine can be defined as follows:
$$P=\pi B Kr_{m}^{2}\Delta rw$$

where

$B$, magnetic loading

$K$, electrical loading

$\Delta r$, radial length of the active area

$r_{m}$, mean magnet radius

* Using **trapezoidal** magnets and coils instead of circular ones are used to increase active length in the magnetic circuit. This makes considerable reduction in diameter of the machine.

The coil emf is calculated as follows:
$$E=\frac{2\pi}{\sqrt2}f\lambda=4.44f\lambda$$

$f$ :frequency 

$\lambda$ :flux linkage

* Wind turbine concept with rectifier and grid connection as a whole, sometimes exposure to over-rated wind speeds. At these times to protect inverter side from breakdown and to provide a safe way to operate the system, extra electrical **dump load** may be employed.
* These resistors can be switched into circuit to maintain dc link voltage constant by dissipating power on themselves and then switched out when the wind speed drops.
* There are two important things about this braking; first one is speed ranges are determined by the dc link voltage limit of the inverter, second one is generator must be rated to carry extra dump load.
